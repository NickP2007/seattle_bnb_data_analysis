```python
import csv
import sqlite3
import pandas as pd
from pathlib import Path

Path('seattle_airbnb.db').touch()
conn = sqlite3.connect('seattle_airbnb.db')
```

```python
c = conn.cursor()

c.execute('''CREATE TABLE listings (id, name, host_id, host_name, neighbourhood_group, neighbourhood, latitude, longitude, room_type, price, minimum_nights, number_of_reviews, last_review, reviews_per_month, calculated_host_listings_count, availability_365
)''')
```

```python
# load the data into a Pandas DataFrame
listings = pd.read_csv('listings.csv')
# write the data to a sqlite table
listings.to_sql('listings', conn, if_exists='append', index = False)

c.execute('''SELECT * FROM listings''').fetchall()
```

```python
c.execute('''CREATE TABLE reviews (listing_id, date
)''')
```

```python
reviews = pd.read_csv('reviews.csv') # load to DataFrame
reviews.to_sql('reviews', conn, if_exists='append', index = False) # write to sqlite table
```

```python
c.execute('''SELECT * FROM reviews''').fetchall() 
```
